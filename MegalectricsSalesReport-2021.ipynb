## 2021 AVERAGE SALES FOR 1ST AND 2ND QUARTER

> We find the average sales for the year 2021 on all stations across the board for the 1st and 2nd quarters. We have observed CLASSIC FM Lagos has an average sales of 29,035,100.00 so far. BEAT FM Lagos has an average sales of 26,847,770.00 so far too. LagosTalks has an average sales of 5,892,667.00 so far. Also, BEAT FM Abuja with the lowest average sales of 589,603.40 so far.

```python
col_2021sales = ['JAN_2021', 'FEB_2021', 'MAR_2021', 'APR_2021','MAY_2021', 'JUNE_2021', 'JULY_2021'];
avg_sales = sales2021.groupby('STATION_NAMES')[col_2021sales].mean()
```

```python
avg_sales.reset_index(level=0, inplace=True)
```

```python
avg_salesnew = avg_sales.mean(axis=1);
```

```python
avg_salesnew = avg_salesnew.to_frame('average_sales');
```

```python
avg_salesnew1 = avg_sales.join(avg_salesnew, lsuffix="_left")
```

```python
avg_salesnew1.drop(['JAN_2021', 'FEB_2021', 'MAR_2021', 'APR_2021','MAY_2021', 'JUNE_2021', 'JULY_2021'], axis = 1, inplace = True);
```

```python
avg_salesnew1
```

```python
avg_salesnew1.plot.bar(x="STATION_NAMES", y="average_sales", figsize = (15,8), color = '#552448')
plt.bar(avg_salesnew1['STATION_NAMES'], avg_salesnew1['average_sales'], color = '#552448', alpha = 0.5);
plt.title('2021 AVERAGE SALES FOR 1ST AND 2ND QUARTERS', fontsize = 20);
plt.xlabel('Station Names', fontsize = 20, labelpad=15);
plt.ylabel('Average Sales', fontsize = 20, labelpad=15);
```
