## 2019 AVERAGE SALES PER YEAR

> We find the average sales for the year 2019 on all stations across the board. We have observed CLASSIC FM Lagos has an average sales of 34,957,240.00. BEAT FM Lagos has an average sales of 28,643,730.00. LagosTalks has an average sales of 7,744,778.00. Also, CLASSIC FM Abuja has the lowest average sales of 632,232.20.

```python
col_2019sales = ['JAN_2019', 'FEB_2019', 'MAR_2019', 'APR_2019','MAY_2019', 'JUNE_2019', 'JULY_2019', 'AUG_2019', 'SEP_2019', 'OCT_2019', 'NOV_2019', 'DEC_2019'];
avg_sales = sales2019.groupby('STATION_NAMES')[col_2019sales].mean()
```

```python
avg_sales.reset_index(level=0, inplace=True)
```

```python
avg_salesnew = avg_sales.mean(axis=1);
```

```python
avg_salesnew = avg_salesnew.to_frame('average_sales');
```

```python
avg_salesnew1 = avg_sales.join(avg_salesnew, lsuffix="_left")
```

```python
avg_salesnew1.drop(['JAN_2019', 'FEB_2019', 'MAR_2019', 'APR_2019','MAY_2019', 'JUNE_2019', 'JULY_2019', 'AUG_2019', 'SEP_2019', 'OCT_2019', 'NOV_2019', 'DEC_2019'], axis = 1, inplace = True);
```

```python
avg_salesnew1
```

```python
avg_salesnew1.plot.bar(x="STATION_NAMES", y="average_sales", figsize = (15,10), color = 'darkslategrey');
plt.bar(avg_salesnew1['STATION_NAMES'], avg_salesnew1['average_sales'], color = 'darkslategrey');
plt.title('2019 AVERAGE SALES PER YEAR', fontsize = 20);
plt.xlabel('Station Names', fontsize = 20, labelpad=15);
plt.ylabel('Average Sales', fontsize = 20, labelpad=15);
```
