```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numbers


%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

## 1ST AND 2ND QUARTERLY COMPARISON SALES 

> We visualized and compears the total sales for the first and second quarter for the years 2019, 2020, and 2021. We did this because we are still in the year 2021. In 2019, 2020, 2021 1st and 2nd we have 43.2%, 30.4%, 26.3% total sales respectively. 

```python
pd.set_option('display.max_columns', 50000)
pd.set_option('display.width', 100000000000000000000)
```

```python
# load in the dataset into a pandas dataframe
firstandsecond_quarter_total_sales = pd.read_csv('YEARLY1STAND2NDQAURTERSALES.csv')
```

```python
firstandsecond_quarter_total_sales
```

```python
my_labels=['1ST AND 2ND QUARTERLY SALES TOTAL 2019','1ST AND 2ND QUARTERLY SALES TOTAL 2020','1ST AND 2ND QUARTERLY SALES TOTAL 2021']
my_colors=['darkred','#486090','royalblue']
explode = (0, 0.2, 0.1) 
firstandsecond_quarter_total_sales.plot.pie(y='1stand2nd_Quarterly_Total-Sales',autopct='%1.1f%%',figsize=(9,9),colors=my_colors,labels=my_labels,explode=explode, shadow=True, pctdistance=0.85, startangle=30);
plt.title('1ST AND 2ND QUARTERLY COMPARISON SALES', fontsize = 20);
plt.xlabel('1st and 2nd Quarterly Sales', fontsize = 15);
plt.ylabel('');
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white');
fig = plt.gcf();
fig.gca().add_artist(centre_circle);
```

## 2019 AND 2020 COMPARISON SALES

> We visualized and compears the total sales for the years 2019, 2020. In 2019, we have 59.3% and in 2020 we have 40.7% total sales.

```python
# load in the dataset into a pandas dataframe
total_yearly_sales= pd.read_csv('YEARLYSALES.csv')
```

```python
total_yearly_sales
```

```python
my_labels=['2019 TOTAL SALES','2020 TOTAL SALES']
my_colors=['#552448','saddlebrown']
explode = (0, 0.2) 
total_yearly_sales.plot.pie(y='Yearly_Total_Sales',autopct='%1.1f%%',figsize=(9,9),labels=my_labels,colors=my_colors,explode=explode,shadow=True, startangle=90);
plt.title('2019 AND 2020 YEAR TO YEAR COMPARISON SALES', fontsize = 20);
plt.xlabel('Year to Year Comparison Sales', fontsize = 15);
plt.ylabel('');
```
